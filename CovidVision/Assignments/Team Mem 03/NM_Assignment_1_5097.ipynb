```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter as co
from math import floor
```

## Loading the dataset

```python
df=pd.read_csv("House Price India.csv")
df.head()
```

## Univariate Analysis

```python
#Countplot

sns.countplot(df['number of floors'][:200])
```

```python
#Boxplot

sns.boxplot(df['Built Year'])
```

```python
#Line plot

plt.plot(df['number of views'][:100])
```

```python
#Histogram

plt.hist(df['number of floors'][:5000])
```

# Bivariate Analysis

```python
#Scatter plot

sns.scatterplot(data=df,x='lot area',y='Price')
```

```python
#Countplot

sns.countplot(data=df[:10],x='Built Year',hue='Number of schools nearby')
```

```python
#Barchart

plt.bar(df['number of views'][:70],df['condition of the house'][:70])
```

```python
#Violin Plot

plt.figure(figsize=(10,5))

plt.violinplot([df['lot area'][:10],df['Price'][:100]],showmedians=True)
```
