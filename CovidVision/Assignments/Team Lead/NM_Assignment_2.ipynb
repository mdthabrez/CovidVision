**Importing Libraries**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras import Input
from tensorflow.keras.layers import Dense
```

**Collecting Dataset**

```python
df = pd.read_csv("/content/House Price India.csv")
df.head()
```

**Data Preprocessing**

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Postal Code'].unique()
```

**Label Encoding**

```python
le = LabelEncoder()
```

```python
df['Postal Code']=le.fit_transform(df['Postal Code'])
```

```python
df['Postal Code'].unique()
```

```python
df.info()
```

```python
x=df.drop('Price',axis=1)
x.head()
```

```python
x.info()
```

```python
y=df['Price']
y
```

****Splitting the Dataset****

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
scaler = MinMaxScaler()

xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

**Defining Model**

```python
model = Sequential()

#input layer

model.add(Input(shape=(22,)))

# hidden layer 1
model.add(Dense(units=32, activation='relu'))

# hidden layer 2
model.add(Dense(units=64, activation='relu'))

# hidden layer 2
model.add(Dense(units=32, activation='relu'))

# Hidden layer 3
model.add(Dense(units=16, activation='relu'))

# output layer
model.add(Dense(units=1, activation='linear'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
input_shape = xtrain.shape  
model.build(input_shape)
model.summary()
```

**Training the model**

```python
model.fit(x=xtrain,y=ytrain.values,batch_size=10,epochs=100,validation_data=(xtest,ytest.values))
```

**Testing the Model**

```python
from sklearn.metrics import r2_score

ypred=model.predict(xtest)
print('R2 score:',r2_score(ytest,ypred))
```

```python
ytest=list(ytest)
ypred=list(ypred)
print('Predicted','Actual')
for i in range(len(ytest)):
    print(ypred[i],ytest[i])
```
